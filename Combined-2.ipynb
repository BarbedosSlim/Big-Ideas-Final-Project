### Building Conditions - Cole Caron
**The Metric**

The Metric I am talking about is Building Condition Ratio. Earthquakes are known for causing high amounts of damage, so the condition of a building matters a lot. Buildings in worse condition before an earthquake are more likely to sustain damage during an earthquake, so it is ideal to be around buildings in better condition. I got this data from Built Environment Conditions 2010, posted by the city of pittsburgh. Even though there are a lot of factors that contribute to a building's condition, it takes a long time for condition to change substancially, so slightly outdated data is still fairly accurate.

## Combining Submatrics

**Implementing a point system to determine the best neighborhood**

### Population Density

* 3000+: -------- 0 points
* 2000-3000: ---- 1 point
* 2000-1500: ---- 2 points
* 1500-1000: ---- 3 points
* 1000-500: ----- 4 points
* 500-0: -------- 5 points

### Building Conditions

* 0.2-: --------- 0 points
* 0.2-0.3: ------ 1 point
* 0.3-0.4: ------ 2 points
* 0.4-0.5: ------ 3 points
* 0.5-0.6: ------ 4 points
* 0.6+: --------- 5 points

### Natural Environment

* 0+: ----------- 0 points
* 0: ------------ 5 points


```python
dataDict = dict()

data_to_copy = top_30_smallest['Persons / sq. mi (2010)'].astype(float)
float_array = data_to_copy.values.astype(float)
poplist = float_array.tolist()
for i, pop_value in enumerate(poplist):
    if pop_value <= 500:
        points = 5
    elif pop_value <= 1000:
        points = 4
    elif pop_value <= 1500:
        points = 3
    elif pop_value <= 2000:
        points = 2
    elif pop_value <= 3000:
        points = 1
    else:
        points = 0
    
    temp = top_30_smallest.iloc[i]['Neighborhood']
    #Calculate points awarded
    if(temp not in dataDict):
        dataDict[temp] = points
    else:
        dataDict[temp] += points

#reset

float_array = buildingData.values.astype(float)
buildlist = float_array.tolist()
for i, build_value in enumerate(buildlist):
    if build_value <= 0.2:
        points = 0
    elif build_value <= 0.3:
        points = 1
    elif build_value <= 0.4:
        points = 2
    elif build_value <= 0.5:
        points = 3
    elif build_value <= 0.6:
        points = 4
    else:
        points = 5
    
    temp = buildingData.index[i]
    if(temp not in dataDict):
        dataDict[temp] = points
    else:
        dataDict[temp] += points
        
#reset

slidelist = Environment1['Landslide Prone (% land area)'].values.astype(float)
for i, slide_value in enumerate(slidelist):
    if slide_value > 0:
        points = 0
    else:
        points = 5
    
    temp = Environment1['Neighborhood'].iloc[i]
    if(temp not in dataDict):
        dataDict[temp] = points
    else:
        dataDict[temp] += points
        
#reset

minelist = Environment2['Undermined (% land area)'].values.astype(float)
for i, mine_value in enumerate(minelist):
    if mine_value > 0:
        points = 0
    else:
        points = 5
        
    temp = Environment2['Neighborhood'].iloc[i]
    if(temp not in dataDict):
        dataDict[temp] = points
    else:
        dataDict[temp] += points
        
scores = pd.Series(dataDict, name = "Fields")
scores = scores.sort_values(ascending=False).head(30)
scores.plot(kind='bar', x='Neighborhood', y='Total Points', legend=None, color=colors)
plt.title("Top 30 Best Neighborhoods for Surviving an Earthquake")
plt.xticks(rotation=90)
plt.ylabel("Total Points")
plt.xlabel("Neighborhood")
fig = plt.gcf()
fig.set_size_inches(13,7)
plt.show()
```
