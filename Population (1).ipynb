The metric I am using is population. I believe population fits very well because housing density in neighborhoods can pose significant problems during earthquakes due to the increased amount of people.
```python
#Import
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
file = 'population-density.csv'
csv_data = pd.read_csv(file)
```

```python
neighborhood_density_csv = csv_data[['Neighborhood', 'Persons / sq. mi (2010)']]
print("CSV Data: Neighborhood and Population Density 2010")
print(neighborhood_density_csv)
```

```python
# Sorting the DataFrame by population density in ascending order
sorted_neighborhoods = neighborhood_density_csv.sort_values(by='Persons / sq. mi (2010)')

# Selecting the top 30 smallest neighborhoods
top_30_smallest = sorted_neighborhoods.head(30)
#print functions
print("Top 30 Smallest Neighborhoods by Population Density in 2010:")
print(top_30_smallest)
```

The people per square mile changes a lot based off of the neighborhood. Many factors can go into this, but it is safe to assume ones with lower numbers are in safer areas. 

```python
# Plotting the top 30 smallest neighborhoods by population density
top_30_smallest.plot(kind='bar', x='Neighborhood', y='Persons / sq. mi (2010)', figsize=(12, 10))
plt.xlabel('Neighborhood') 
plt.ylabel('Population Density (Persons / sq. mi)')
plt.title('Top 30 Smallest Neighborhoods by Population Density in 2010')
plt.xticks(rotation=90)  
plt.tight_layout()  
plt.show()
```

The data clearly shows that certain neighborhoods are better fit to survive an earthquake than others.

**Landslide Prone Area**

I choose this data point because landslide are common after earthquakes. I wanted to make sure the neighborhood I choose was not prone to have landslide damage after the earthquake.

```python
import pandas as pd
import matplotlib.pyplot as plt
Environment = pd.read_csv("Environment.csv")
Environment['Landslide Prone (% land area)'] = pd.to_numeric(Environment['Landslide Prone (% land area)'].str.rstrip('%'), errors='coerce')
Environment = Environment.dropna(subset=['Landslide Prone (% land area)'])
Environment = Environment.sort_values(by='Landslide Prone (% land area)')
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown', 'pink', 'gray', 'olive', 'cyan',
              'lightcoral', 'darkorange', 'limegreen', 'darkred', 'darkviolet', 'saddlebrown', 'hotpink', 'dimgray',
              'darkkhaki', 'teal', 'lightblue', 'goldenrod', 'darkgreen', 'darkmagenta', 'peru', 'indigo', 'sienna',
              'salmon', 'steelblue', 'tomato']
Environment = Environment.head(30)
Environment.plot(kind='bar', x='Neighborhood', y='Landslide Prone (% land area)', legend=None,color=colors)
plt.title('Landslide Prone Area by Neighborhood (Lowest 30)')
plt.xlabel('Neighborhood')
plt.ylabel('Landslide Prone (% land area)')
plt.xticks(rotation=90)  
fig = plt.gcf()
fig.set_size_inches(13, 7)
plt.tight_layout()
plt.show()
```


There were several places that had 0% landslide prone areas. There were 14 places that had 0%. Places such as Allegheny Center, East Hills, East Carnegie and ect...

**Undermined Area**

I choose this data point because undermined area can have more damage caused by earthquake. Since the underground has been mined it is less stable then a place that hasn't been mined.

```python
import pandas as pd
import matplotlib.pyplot as plt
Environment = pd.read_csv("Environment.csv")
Environment['Undermined (% land area)'] = pd.to_numeric(Environment['Undermined (% land area)'].str.rstrip('%'), errors='coerce')
Environment = Environment.dropna(subset=['Undermined (% land area)'])
Environment = Environment.sort_values(by='Undermined (% land area)')
Environment = Environment.head(50)
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown', 'pink', 'gray', 'olive', 'cyan',
          'lightcoral', 'darkorange', 'limegreen', 'darkred', 'darkviolet', 'saddlebrown', 'hotpink', 'dimgray',
          'darkkhaki', 'teal', 'lightblue', 'goldenrod', 'darkgreen', 'darkmagenta', 'peru', 'indigo', 'sienna',
          'salmon', 'steelblue', 'tomato']
Environment.plot(kind='bar', x='Neighborhood', y='Undermined (% land area)', legend=None, color=colors)
plt.title('Undermined Area by Neighborhood (Lowest 50)')
plt.xlabel('Neighborhood')
plt.ylabel('Undermined (% land area)')
plt.xticks(rotation=90)  
fig = plt.gcf()
fig.set_size_inches(13, 7)
plt.tight_layout()
plt.show()
```

There were several places that had 0% undermined areas. There were 41 places that had 0%. Places such as Allegheny Center, Fairywood, Chateau and ect...
